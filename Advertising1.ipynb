#  statistical analysis

```python
import statsmodels.formula.api as smf
lm=smf.ols(formula='sales ~ TV',data=df).fit()
lm.summary()
```

```python
lm=smf.ols(formula='sales ~ TV+radio',data=df).fit()
lm.summary()
```

```python
lm=smf.ols(formula='sales ~ TV+radio+newspaper',data=df).fit()
lm.summary()
```
